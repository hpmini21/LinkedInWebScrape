# Text Scraping 1 - Using HTML

## BeautifulSoup

## Crawling (banyak pages)

Algorithm:

1. tentukan alamat url yang akan di store (tanpa page number)
2. buat array kosong sesuai dengan jumlah feature yang akan dimasukkan
3. buat loop dari page yang akan diambil
4. append terhadap array dari setiap tag



```python
from bs4 import BeautifulSoup
import requests

root = 'https://www.linkedin.com/jobs/search/?distance=25.0&geoId=102478259&keywords=data%20analysis&origin=HISTORY&start='

job_name = []
loc = []
company = []
jobLink = []

for page in range(0, 250, 25):
  url = requests.get(root + str(page))
  soup = BeautifulSoup(url.text, 'html.parser')

  for q in soup.find_all('span', {'class': 'sr-only'}):
    if q.text.strip() == 'LinkedIn' or q.text.strip() == 'Expand search':
      continue
    job_name.append(q.text.strip())

  for a in soup.find_all("span", {"class":"job-search-card__location"}):
    loc.append(a.text.strip())

  for t in soup.find_all('a', {"class" :"hidden-nested-link"}):
    company.append(t.text.strip())

  for z in soup.find_all('a'):
    href = z.get('href')  # Ambil nilai atribut href
    if href and "linkedin.com/jobs/view/" in href:
      jobLink.append(href)
```

```python
len(job_name), len(loc), len(company),len(jobLink)
```

```python
import pandas as pd
df = pd.DataFrame(
    {'jobname':job_name,
     'location':loc,
     'company':company,
     'jobLink':jobLink
     })
df
# df['jobLink'][0]

```
